```python
from PIL import Image, ImageOps
import os

def invert_image_colors(image_path):
    # Load the image
    image = Image.open(image_path)
    
    # Invert the colors
    inverted_image = ImageOps.invert(image)
    
    return inverted_image

def process_directory(input_directory, output_directory):
    # Walk through the input directory tree
    for root, dirs, files in os.walk(input_directory):
        for file in files:
            if file.endswith(".jpg") or file.endswith(".png"):
                # Construct the full file path
                input_path = os.path.join(root, file)
                
                # Construct the corresponding output path
                relative_path = os.path.relpath(root, input_directory)
                output_path_root = os.path.join(output_directory, relative_path)
                if not os.path.exists(output_path_root):
                    os.makedirs(output_path_root)
                output_path = os.path.join(output_path_root, file)
                
                # Invert the image colors and save the result
                inverted_image = invert_image_colors(input_path)
                inverted_image.save(output_path)

# Specify the path to your directory and output directory
input_directory = 'digit_images_sing'
output_directory = 'converted_new_sing'
process_directory(input_directory, output_directory)
```

```python
import os
import torch
from torchvision import datasets, transforms
from PIL import Image
import numpy as np

# Load MNIST dataset
transform = transforms.Compose([transforms.ToTensor()])
mnist = datasets.MNIST(root='./data', train=True, download=True, transform=transform)

# Function to get images for the digits from the dataset
def get_images_for_digits(dataset, digit1, digit2):
    # Extract images for the digits
    indices1 = [i for i, (img, label) in enumerate(dataset) if label == digit1]
    indices2 = [i for i, (img, label) in enumerate(dataset) if label == digit2]
    
    # Randomly select index for the digits' images
    idx1 = np.random.choice(indices1)
    idx2 = np.random.choice(indices2)

    # Get the images for these indices
    img1, _ = dataset[idx1]
    img2, _ = dataset[idx2]

    # Create a slash fraction bar using torch.ones with the same width and height as the original images
    slash_fraction_bar = torch.ones([1, img1.shape[1], img1.shape[2]])

    return img1, img2, slash_fraction_bar

# Generate 30 fraction images for all combinations of digits from 0 to 9
for num in range(10):
    for den in range(10):
        # Create a new directory for each fraction
        os.makedirs(f'result_fraction/{num}_{den}', exist_ok=True)
        
        for i in range(91, 111):  # Generate images from 31 to 60
            # Create new images
            nominator_img, denominator_img, slash_fraction_bar_img = get_images_for_digits(mnist, num, den)

            # Convert the images to PIL Images and save them
            Image.fromarray(nominator_img.numpy()[0] * 255).convert('L').save(f'numerator/{num}_{i}.png')
            Image.fromarray(denominator_img.numpy()[0] * 255).convert('L').save(f'denominator/{den}_{i}.png')
            
            # Create the directory if it does not exist
            os.makedirs('slash_fraction_bar', exist_ok=True)
            
            Image.fromarray(slash_fraction_bar_img.numpy()[0] * 255).convert('L').save(f'slash_fraction_bar/slash_fraction_bar_{i}.png')

            # Adjust the dimensions of the slash fraction bar tensor
            expanded_slash_fraction_bar = slash_fraction_bar_img.expand(-1, nominator_img.shape[1], nominator_img.shape[2])

            # Create the final image by concatenating the images horizontally with a slash fraction bar in the middle
            final_img = torch.cat((nominator_img, expanded_slash_fraction_bar, denominator_img), dim=2)

            # Convert the final image to a PIL Image and save it
            Image.fromarray(final_img.numpy()[0] * 255).convert('L').save(f'result_fraction/{num}_{den}/{num}_{den}_{i}.png')

            print(f'Generated fraction {i}: {num}/{den}')
```

```python
#two digits fraction
import random
import os
from PIL import Image, ImageDraw, ImageFont

# Step 1: Load MNIST dataset
# (Assuming you have already downloaded and loaded the MNIST dataset)

# Step 2: Generate fraction pairs
num_pairs = 100  # Specify the number of fraction pairs to generate

fraction_pairs = []
for _ in range(num_pairs):
    numerator = random.randint(10, 99)  # Generate two-digit numerator
    denominator = random.randint(10, 99)  # Generate two-digit denominator
    fraction_pairs.append((numerator, denominator))

# Step 3: Create fraction images
image_width = 128
image_height = 128

font_size = 40
font = ImageFont.truetype("arial.ttf", font_size)

fraction_images = []
for numerator, denominator in fraction_pairs:
    image = Image.new("RGB", (image_width, image_height), (255, 255, 255))
    draw = ImageDraw.Draw(image)

    numerator_position = (image_width // 2, (image_height - font_size) // 2)
    denominator_position = (image_width // 2, (image_height + font_size) // 2)

    draw.text(numerator_position, str(numerator), fill=(0, 0, 0), anchor="mm", font=font)
    draw.text(denominator_position, str(denominator), fill=(0, 0, 0), anchor="mm", font=font)
    
    # Draw the fraction bar
    fraction_bar_position = (image_width // 2, image_height // 2)
    fraction_bar_length = image_width // 2
    fraction_bar_thickness = 2
    draw.line([(fraction_bar_position[0] - fraction_bar_length // 2, fraction_bar_position[1]), 
               (fraction_bar_position[0] + fraction_bar_length // 2, fraction_bar_position[1])],
              fill=(0, 0, 0), width=fraction_bar_thickness)

    fraction_images.append(image)

# Step 4: Save the fraction images
output_directory = "fraction_images/"

# Create the output directory if it doesn't exist
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

for i, image in enumerate(fraction_images):
    image.save(output_directory + f"fraction_{i}.png")
```

```python
#The right code for the two digits fraction
import random
import os
from PIL import Image, ImageDraw
from torchvision.datasets import MNIST
from torchvision.transforms import ToTensor, Resize, ToPILImage

# Step 1: Load MNIST dataset
mnist_dataset = MNIST(root=".", train=True, download=True, transform=ToTensor())

# Step 2: Generate fraction pairs
num_pairs = 100  # Specify the number of fraction pairs to generate

fraction_pairs = []
for _ in range(num_pairs):
    numerator_digits = random.sample(range(10), 2)  # Randomly select two digits for the numerator
    denominator_digits = random.sample(range(10), 2)  # Randomly select two digits for the denominator
    
    numerator_images = []
    denominator_images = []
    
    # Render numerator digits
    for digit in numerator_digits:
        digit_index = random.randint(0, len(mnist_dataset) - 1)  # Randomly select an index for the digit image
        digit_image = mnist_dataset[digit_index][0]  # Access the image at the selected index for the digit
        
        numerator_images.append(digit_image)
    
    # Render denominator digits
    for digit in denominator_digits:
        digit_index = random.randint(0, len(mnist_dataset) - 1)  # Randomly select an index for the digit image
        digit_image = mnist_dataset[digit_index][0]  # Access the image at the selected index for the digit
        
        denominator_images.append(digit_image)
    
    fraction_pairs.append((numerator_images, denominator_images))

# Step 3: Create fraction images
image_width = 128
image_height = 128

fraction_images = []
for numerator_images, denominator_images in fraction_pairs:
    numerator_images_resized = []
    denominator_images_resized = []
    
    # Resize numerator digits
    for digit_image in numerator_images:
        digit_image_resized = Resize((image_width // 2, image_height // 2))(digit_image)
        numerator_images_resized.append(digit_image_resized)
    
    # Resize denominator digits
    for digit_image in denominator_images:
        digit_image_resized = Resize((image_width // 2, image_height // 2))(digit_image)
        denominator_images_resized.append(digit_image_resized)
    
    # Create the fraction image
    image = Image.new("RGB", (image_width, image_height), (255, 255, 255))
    
    # Calculate the height of the digits
    digit_height = numerator_images_resized[0].shape[1]
    
    # Calculate the y-coordinate for placing the digits
    digits_y = (image_height - digit_height * 2) // 2
    
    # Paste numerator digits
    x_offset = (image_width // 2 - numerator_images_resized[0].shape[2]) // 2
    for digit_image in numerator_images_resized:
        digit_pil_image = ToPILImage()(digit_image)
        width, height = digit_pil_image.size
        image.paste(digit_pil_image, (x_offset, digits_y, x_offset + width, digits_y + height))
        x_offset += width
    
    # Paste denominator digits
    x_offset = (image_width // 2 - denominator_images_resized[0].shape[2]) // 2
    for digit_image in denominator_images_resized:
        digit_pil_image = ToPILImage()(digit_image)
        width, height = digit_pil_image.size
        image.paste(digit_pil_image, (x_offset, digits_y + digit_height, x_offset + width, digits_y + digit_height + height))
        x_offset += width
    
    # Draw the fraction bar
    fraction_bar_y = image_height // 2  # Position the fraction bar exactly in the middle
    fraction_bar_height = 2
    fraction_bar_x_start = 0
    fraction_bar_x_end = image_width
    draw = ImageDraw.Draw(image)
    draw.rectangle([(fraction_bar_x_start, fraction_bar_y - fraction_bar_height // 2), (fraction_bar_x_end, fraction_bar_y + fraction_bar_height // 2)], fill=(255, 255, 255))
    
    fraction_images.append(image)

# Step 4: Save the fraction images
output_directory = "fraction_imageZ/"

# Create the output directory if it doesn't exist
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

for i, image in enumerate(fraction_images):
    image.save(output_directory + f"fraction_{i}.png")
```

```python
#generate two digits
import random
import os
from PIL import Image, ImageDraw
from torchvision.datasets import MNIST
from torchvision.transforms import ToTensor, Resize, ToPILImage

# Step 1: Load MNIST dataset
mnist_dataset = MNIST(root=".", train=False, download=True, transform=ToTensor())

# Step 2: Generate number images with concatenated digits
num_images = 500
image_width = 128
image_height = 64
digit_width = image_width // 2
digit_height = image_height

number_images = []

for _ in range(num_images):
    tens_digit = random.randint(1, 19)  # Randomly select a tens digit between 1 and 19
    ones_digit = random.randint(0, 9)  # Randomly select an ones digit between 0 and 9

    number = tens_digit * 10 + ones_digit  # Concatenate the tens and ones digits to form the number

    tens_digit_image = mnist_dataset[tens_digit][0]  # Get the image for the tens digit
    ones_digit_image = mnist_dataset[ones_digit][0]  # Get the image for the ones digit

    # Convert tensor images to PIL images
    tens_digit_image_pil = ToPILImage()(tens_digit_image)
    ones_digit_image_pil = ToPILImage()(ones_digit_image)

    # Resize digit images
    tens_digit_image_resized = Resize((digit_width, digit_height))(tens_digit_image_pil)
    ones_digit_image_resized = Resize((digit_width, digit_height))(ones_digit_image_pil)

    # Create a new image with the concatenated digits
    image = Image.new("RGB", (image_width, image_height))
    image.paste(tens_digit_image_resized, (0, 0, digit_width, digit_height))
    image.paste(ones_digit_image_resized, (digit_width, 0, image_width, digit_height))

    number_images.append(image)

# Step 3: Save the number images
output_directory = "number_images_concatenated_new/"

# Create the output directory if it doesn't exist
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

for i, image in enumerate(number_images):
    image.save(output_directory + f"number_{i}.png")
```
